```python
import nltk, os
import pandas as pd
from nltk.tokenize import RegexpTokenizer
```

```python
positive_score_list = []
negative_score_list = []
polarity_score_list = []
average_sentence_length_list = []
percentage_of_complex_words_list = []
fog_index_list = []
complex_word_count_list = []
word_count_list = []
uncertainty_score_list = []
constraining_score_list = []
positive_word_proportion_list = []
negative_word_proportion_list = []
uncertainty_word_proportion_list = []
constraining_word_proportion_list = []

folder = "data/"
stop_words = nltk.corpus.stopwords.words("english")

sentiments_df = pd.read_csv("sentiment_dictionary.csv")
pos_series = sentiments_df["Positive"].dropna()
neg_series = sentiments_df["Negative"].dropna()
positive_words_dictionary = pos_series.str.lower()
negative_words_dictionary = neg_series.str.lower()

print("len(positive_words_dictionary) = ", len(positive_words_dictionary))
print("len(negative_words_dictionary = ", len(negative_words_dictionary))

df = pd.read_csv("LoughranMcDonald_MasterDictionary_2018.csv")

# Making Complex Words Dictionary

complex_words = df["Word"].loc[df["Syllables"]>2]
complex_words_dictionary = complex_words.str.lower()
print("len(complex_words_dictionary) = ", len(complex_words_dictionary))

# Making uncertainty Words Dictionary

uncertainty_df = pd.read_excel("uncertainty_dictionary.xlsx")
uncertainty_words_series = uncertainty_df["Word"].str.lower()
print("len(uncertainty_words_series) = ",len(uncertainty_words_series))

# Making constraining Words Dictionary

constraining_df = pd.read_excel("constraining_dictionary.xlsx")
constraining_words_series = constraining_df["Word"].str.lower()
print("len(constraining_words_series) = ",len(constraining_words_series))

# Reading Data

cik_df = pd.read_excel("cik_list.xlsx")
file_series = cik_df["SECFNAME"]
for file in file_series:
    file_name = os.path.split(file)[-1]
    with open("data/%s" %file_name) as file_obj:
        raw_data = file_obj.read()
    #     print(raw_data)

        corpusReader = nltk.corpus.PlaintextCorpusReader(folder, ['%s' %file_name])
        sentence_count = len(corpusReader.sents())
        print ("The number of sentences =", sentence_count)

        # Tokenising Raw Data

        tokenizer = RegexpTokenizer(r'[A-Za-z]{2,}')
        tokens = tokenizer.tokenize(raw_data)
        lowered_tokens = [t.lower() for t in tokens]
        lowered_tokens = [w for w in lowered_tokens if not w in stop_words]
        lowered_tokens_series = pd.Series(lowered_tokens)
        
        # Word Count
        word_count = len(lowered_tokens)
        print("word_count = ", word_count)
        print("len(lowered_tokens) = ",len(lowered_tokens))

        #Average Sentence Length
        average_sentence_length = (word_count/sentence_count)
        print ("average_sentence_length =", average_sentence_length)

        # Finding Complex Words in Data

        complex_word_count = len(lowered_tokens_series[lowered_tokens_series.isin(complex_words_dictionary)])
        print("complex_word_count = ", complex_word_count)

        percentage_of_complex_words = (complex_word_count/word_count)*100
        print("percentage_of_complex_words = ", percentage_of_complex_words)
        
        # Fog Index
        
        fog_index = 0.4 * (average_sentence_length + percentage_of_complex_words)
        print("fog_index = ", fog_index)

        # Finding Positive Negative Score

        positive_score = 0
        negative_score = 0

        positive_score = len(lowered_tokens_series[lowered_tokens_series.isin(positive_words_dictionary)])
        negative_score = len(lowered_tokens_series[lowered_tokens_series.isin(negative_words_dictionary)])

        print("positive_score = ",positive_score)
        print("negative_score = ",negative_score)

        # Finding Polarity Score

        polarity_score = (positive_score-negative_score)/((positive_score+negative_score)+0.000001)
        print("polarity_score = ",polarity_score)

        # Finding subjectivity Score

        subjectivity_score = (positive_score+negative_score)/((len(lowered_tokens))+0.000001)
        print("subjectivity_score = ",subjectivity_score)

        # Finding Uncertainty Score

        uncertainty_score = len(lowered_tokens_series[lowered_tokens_series.isin(uncertainty_words_series)])
        print("uncertainty_score = ",uncertainty_score)

        # Finding Constraining Score

        constraining_score = len(lowered_tokens_series[lowered_tokens_series.isin(constraining_words_series)])
        print("constraining_score = ",constraining_score)

        # Finding Positive and Negative Word Proportion

        positive_word_proportion = (positive_score/word_count)
        print("positive_word_proportion = ",positive_word_proportion)

        negative_word_proportion = (negative_score/word_count)
        print("negative_word_proportion = ",negative_word_proportion)

        # Uncertainty Word Proportion

        uncertainty_word_proportion = (uncertainty_score/word_count)
        print("uncertainty_word_proportion = ",uncertainty_word_proportion)

        # Constraining Word Proportion

        constraining_word_proportion = (constraining_score/word_count)
        print("constraining_word_proportion = ",constraining_word_proportion, "\n")
        
        positive_score_list.append(positive_score)
        negative_score_list.append(negative_score)
        polarity_score_list.append(polarity_score)
        average_sentence_length_list.append(average_sentence_length)
        percentage_of_complex_words_list.append(percentage_of_complex_words)
        fog_index_list.append(fog_index)
        complex_word_count_list.append(complex_word_count)
        word_count_list.append(word_count)
        uncertainty_score_list.append(uncertainty_score)
        constraining_score_list.append(constraining_score)
        positive_word_proportion_list.append(positive_word_proportion)
        negative_word_proportion_list.append(negative_word_proportion)
        uncertainty_word_proportion_list.append(uncertainty_word_proportion)
        constraining_word_proportion_list.append(constraining_word_proportion)
```

# Writing Results

```python
# cik_df
cik_df["positive_score"] = positive_score_list
cik_df["negative_score"] = negative_score_list
cik_df["polarity_score"] = polarity_score_list
cik_df["average_sentence_length"] = average_sentence_length_list
cik_df["percentage_of_complex_words"] = percentage_of_complex_words_list
cik_df["fog_index"] = fog_index_list
cik_df["complex_word_count"] = complex_word_count_list
cik_df["word_count"] = word_count_list
cik_df["uncertainty_score"] = uncertainty_score_list
cik_df["constraining_score"] = constraining_score_list
cik_df["positive_word_proportion"] = positive_word_proportion_list
cik_df["negative_word_proportion"] = negative_word_proportion_list
cik_df["uncertainty_word_proportion"] = uncertainty_word_proportion_list
cik_df["constraining_word_proportion"] = constraining_word_proportion_list
cik_df.to_csv("result.csv")
```
